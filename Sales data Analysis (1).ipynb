```python
def clean_sales_data(df):
    original_shape=df.shape
    cleaning_log=[]
    df=df.copy()
    #handle missing values
    missing_before=df.isnull().sum()
    
    #filling num values with appro value
    numeric_cols=['discounted_price','actual_price','discount_percentage','rating','rating_count']
    for cols in numeric_cols:
        if cols in df.columns:
            #using median for missings values of prices and ratings
            df[cols]=df[cols].fillna(df[cols].median())
            
    #filling missing categorical values
    categorical_cols=['category','product_name','about_product']
    for cols in categorical_cols:
        if cols in df.columns:
            df[cols]=df[cols].fillna('Unknown')
print(df)
```

```python
missing_after=df.isnull().sum()
print(missing_after)
```

```python
#removing duplicates
duplicates_count=df.duplicated().sum()
print(duplicates_count)
```

```python
#clean and standardise text cols
text_cols=['product_name','category']
for cols in text_cols:
    if cols in df.columns:
        df[cols]=df[cols].str.strip()
        df[cols]=df[cols].str.title()
    print(df)
```

```python
rating_cols=['rating','rating_count']
for cols in rating_cols:
    if cols in df.columns:
        df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
        df['rating'] = df['rating'].clip(0.0, 5.0)
    print(df[cols])
```

```python
if 'discount_percentage' in df.columns:
    # Ensure discount percentage is between 0 and 100
    df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
    df['discount_percentage'] = df['discount_percentage'].clip(0, 100)
    print(df['discount_percentage'])
```

```python
price_cols = ['discounted_price', 'actual_price']
for cols in price_cols:
    if cols in df.columns:
        df[cols] = df[cols].replace('[₹,]', '', regex=True)
        df[cols] = df[cols].astype(int)    
    print(df[cols])
```

```python
# Calculate the discount percentage
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100
print(df['discount_percentage'])
```

```python
# Save CSV file to a specific folder
df.to_csv('C:\\Users\\91939\\Downloads\\archive (5)\\cleaned_sales_dataset.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:\\Users\\91939\\Downloads\\archive (5)\\cleaned_sales_dataset.csv')
```

```python
# Calculate the discount percentage
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100
print(df['discount_percentage'])
```

```python
#grouping discount to intervals
df['discount_range']=pd.cut(df['discount_percentage'],
                            bins=[0,10,20,30,40,50,100],
                            labels=['0-10%','11-20%','21-30%','31-40%','41-50%','50%+'])
print(df['discount_range'])
```

```python
#visualisation of intervals of discount
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='discount_percentage',bins=30)
plt.title('distribution of discount percentages')
plt.xlabel('Discount percentage')
plt.ylabel('Number of Products')
plt.show()
```

```python
# 4. Average discount by product category
category_discount = df.groupby('category')['discount_percentage'].mean().reset_index()
print('\nAverage Discount by Category:')
print(category_discount)
```

```python
#visualisation of discount by product category
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='category', y='discount_percentage')
plt.title('distribution of categories')
plt.xlabel('categories')
plt.ylabel('discount percentage')
plt.xticks(rotation=45,fontsize=5)
plt.show()
```

```python
#Discount percentage vs. actual price
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='discount_percentage', y='actual_price')
plt.title('Discount percentage vs. actual price')
plt.xlabel('Discount percentage')
plt.ylabel('actual prices')
plt.show()
```

```python
#Discount vs Rating Correlation
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='discount_percentage', y='rating_count', 
                   size='rating_count', sizes=(20, 200), alpha=0.6)
plt.title('Discount Percentage vs Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating count')
plt.show()
```

```python
#Discounts vs. Ratings
plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='discount_percentage', data=df)
plt.title('Discount Percentage vs. Product Rating')
plt.xlabel('Product Rating')
plt.ylabel('Discount Percentage')
plt.show()
```
